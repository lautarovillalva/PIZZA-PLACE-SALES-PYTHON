```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pizzas = pd.read_csv("pizza_sales/pizzas.csv")
pizza_types = pd.read_csv("pizza_sales/pizza_types.csv", encoding = 'unicode_escape')
orders = pd.read_csv("pizza_sales/orders.csv")
order_details = pd.read_csv("pizza_sales/order_details.csv")
```

```python
orders_full = orders.join(order_details.set_index("order_id"), on="order_id")
orders_full
```

```python
# df.join(other.set_index('key'), on='key')
pizzas_full = pizzas.join(pizza_types.set_index("pizza_type_id"), on="pizza_type_id")
pizzas_full
```

```python
df = orders_full.merge(pizzas_full, on="pizza_id").sort_values(by="order_id")
df
```

```python
print(f"Registros: {df.shape[0]} | Columnas: {df.shape[1]}")
```

```python
df.groupby(["order_id"])["quantity"].sum().sort_values(ascending=False)
```

```python
id_bo = df.groupby(["order_id"])["quantity"].sum().sort_values(ascending=False).index[0]
id_bo
```

```python
bigger_order = df.loc[df["order_id"]==id_bo, :].sort_values(["category", "quantity"])
bigger_order
```

```python
total_price = (bigger_order["quantity"]*bigger_order["price"]).sum()
total_quantity = bigger_order["quantity"].sum() 

print(f"El pedido mas grande es el {id_bo} con {total_quantity} pizzas y ${total_price}")
```
